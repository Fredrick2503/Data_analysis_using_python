```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("bangalore-rainfall-data-1900-2024-sept.csv")

# Filter data for the years 1999 to 2024
data_filtered = data[(data['Year'] >= '1999') & (data['Year'] <= '2024')]

# Find the month with the highest rainfall for each year
data_filtered['Max Month'] = data_filtered[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']].idxmax(axis=1)
data_filtered['Max Rainfall'] = data_filtered[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']].max(axis=1)

# Plot
plt.figure(figsize=(14, 8))
plt.scatter(data_filtered['Year'], data_filtered['Max Rainfall'], c='b', marker='o')
for i in range(len(data_filtered)):
    plt.text(data_filtered['Year'].iloc[i], data_filtered['Max Rainfall'].iloc[i], data_filtered['Max Month'].iloc[i], fontsize=9)
plt.title('Month of Highest Rainfall for Each Year (1999-2024)')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("bangalore-rainfall-data-1900-2024-sept.csv")

# Filter data for the years 1999 to 2023
data_filtered = data[(data['Year'] >= '1999') & (data['Year'] <= '2023')]

# Calculate the average rainfall
average_rainfall = data_filtered['Total'].mean()

# Plot
plt.figure(figsize=(14, 8))
plt.bar(data_filtered['Year'], data_filtered['Total'], color='blue', edgecolor='black', label='Total Rainfall')
plt.axhline(y=average_rainfall, color='r', linestyle='--', label='Average Rainfall')
plt.title('Total Rainfall (1999-2023) with Average Rainfall')
plt.xlabel('Year')
plt.ylabel('Total Rainfall (mm)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

Based on the rainfall data from 1999 to 2023, several key insights can be drawn. The total rainfall for each year shows significant variability, with some years experiencing much higher rainfall than others. For instance, 2022 recorded an exceptionally high total rainfall of 1957.7 mm, while 2003 had a much lower total of 678.7 mm. This variability can be attributed to climatic phenomena such as El NiÃ±o and La NiÃ±a, which have a notable impact on rainfall patterns. El NiÃ±o years, like 2023, tend to have lower rainfall, while La NiÃ±a years, such as 2022, often see higher rainfall.

The bar chart visualizing the total rainfall for each year between 1999 and 2023, with a line indicating the average rainfall, provides a clear picture of these fluctuations. The average rainfall over this period is approximately 1077.1 mm, as indicated by the red dashed line. Years with rainfall significantly above or below this average are easily identifiable, highlighting the impact of climatic events.

Additionally, the scatter plot showing the month with the highest rainfall for each year between 1999 and 2024 reveals interesting patterns. For example, May and September are frequently the months with the highest rainfall, indicating the influence of the monsoon season. This information is crucial for understanding seasonal rainfall distribution and planning for water resource management.

Overall, these visualizations and analyses provide valuable insights into the rainfall patterns in Bangalore, helping to understand long-term trends and the impact of climatic phenomena on annual rainfall.
